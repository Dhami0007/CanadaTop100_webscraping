```python
from bs4 import BeautifulSoup
import requests
import re
import ast
import pandas as pd
```

```python
url = "https://www.canadastop100.com/national/"
```

```python
page = requests.get(url)
```

```python
soup = BeautifulSoup(page.text, "html.parser")
print(soup.prettify())
```

Here is the whole html of the requested page

```python
script_tag = soup.find("script", text=re.compile(r"const winnerData = "))
```

```python
content = str(script_tag)
print(content)
```

We have the script tag of the html now. Let us extract the list of the companies from it.

```python
match = re.search(r"const winnerData = (\[.*?\]);", content, re.DOTALL)
```

```python
winner_data = match.group(1)
print(winner_data)
```

We have the list now!

```python
my_list = ast.literal_eval(winner_data)
```

Lets inspect an element of the list 

```python
# Type of the elements in the list
print(type(my_list[0]))

# element format
for key, value in my_list[0].items():
    print(f"{key}: {value}")
```
